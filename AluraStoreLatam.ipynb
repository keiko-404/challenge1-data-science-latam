### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación
En este debes calcular la cantidad de productos vendidos por categoría en cada tienda. La idea es agrupar los datos por categoría y contar el número de ventas de cada tipo, mostrando las categorías más populares de cada tienda.


```python
# suma de la columna Precio para sacar el precio total de la tienda 1
ingreso_total_t1 = tienda["Precio"].sum()
print(f'El valor total de la tienda 1 es : ${ingreso_total_t1}')
```

```python
ingreso_total_t2 = tienda2["Precio"].sum()
print(f'El valor total de la tienda 2 es : ${ingreso_total_t2}')
```

```python
ingreso_total_t3 = tienda3["Precio"].sum()
print(f'El valor total de la tienda 3 es : ${ingreso_total_t3}')
```

```python
ingreso_total_t4 = tienda4["Precio"].sum()
print(f'El valor total de la tienda 4 es : ${ingreso_total_t4}')
```

# 2. Ventas por categoría
En este debes calcular la cantidad de productos vendidos por categoría en cada tienda. La idea es agrupar los datos por categoría y contar el número de ventas de cada tipo, mostrando las categorías más populares de cada tienda.

```python
tienda['Categoría del Producto'].unique()
```

```python
#cantidad de productos vendidos por cateogria
cpv_tienda = tienda['Categoría del Producto'].value_counts()
cpv_tienda2 = tienda2['Categoría del Producto'].value_counts()
cpv_tienda3 = tienda3['Categoría del Producto'].value_counts()
cpv_tienda4 = tienda4['Categoría del Producto'].value_counts()


cpv_tienda
```

```python
# se agrupan las filas y suma el precio para sacar el monto total vendido por categoria
# total vendio por categoria
tvc_tienda = tienda.groupby('Categoría del Producto')['Precio'].sum()
tvc_tienda2 = tienda2.groupby('Categoría del Producto')['Precio'].sum()
tvc_tienda3 = tienda3.groupby('Categoría del Producto')['Precio'].sum()
tvc_tienda4 = tienda4.groupby('Categoría del Producto')['Precio'].sum()
tvc_tienda
```

```python
#categorias mas populares con sort
cmp_tienda = tienda.groupby('Categoría del Producto')[['Precio']].sum().sort_values('Precio', ascending=False)
cmp_tienda2 = tienda2.groupby('Categoría del Producto')[['Precio']].sum().sort_values('Precio', ascending=False)
cmp_tienda3 = tienda3.groupby('Categoría del Producto')[['Precio']].sum().sort_values('Precio', ascending=False)
cmp_tienda4 = tienda4.groupby('Categoría del Producto')[['Precio']].sum().sort_values('Precio', ascending=False)
cmp_tienda
```

# 3. Calificación promedio de la tienda
En este paso, debes calcular las calificaciones promedio de los clientes para cada tienda. El objetivo es conocer la satisfacción del cliente con los productos vendidos.

```python
prom_tienda = tienda['Calificación'].mean()
prom_tienda2 = tienda2['Calificación'].mean()
prom_tienda3 = tienda3['Calificación'].mean()
prom_tienda4 = tienda4['Calificación'].mean()

print(f'La calificación promedio de la tienda 1 es: {prom_tienda:.2f}')
print(f'La calificación promedio de la tienda 2 es: {prom_tienda2:.2f}')
print(f'La calificación promedio de la tienda 3 es: {prom_tienda3:.2f}')
print(f'La calificación promedio de la tienda 4 es: {prom_tienda4:.2f}')
```

# 4. Productos más y menos vendidos
En este paso, debes identificar qué productos fueron los más vendidos y los menos vendidos en cada tienda. Visualiza los resultados para que quede claro qué productos destacaron en ventas en cada tienda.

```python
#TIENDA 1
def top_mas_ingresos(df, tienda):
    # agrupar por producto: cantidad, precio unitario, ingreso total
    resumen = df.groupby('Producto').agg(
        cantidad_vendida=('Producto', 'count'),
        ingreso_total=('Precio', 'sum')
    )

    # top 3 que generaron mas dinero
    top_mas = resumen.sort_values('ingreso_total', ascending=False).head(3)

    print('-- TIENDA- --')
    print('top 3 que generaron mas dinero')
    print(top_mas)
    return top_mas

def top_menos_ingresos(df, tienda):

    # agrupar por producto
    resumen = df.groupby('Producto').agg(
        cantidad_vendida=('Producto', 'count'),
        ingreso_total=('Precio', 'sum')
    )

    # top 3 que generaron men os dinero
    top_menos = resumen.sort_values('ingreso_total', ascending=True).head(3)

    print('\ntop 3 que generaron menos dinero')
    print(top_menos)

top_mas_ingresos(tienda, 'Tienda 1')
top_menos_ingresos(tienda, 'Tienda 1')
```

```python
cmp_tienda = tienda.groupby('Categoría del Producto')[['Precio']].sum().sort_values('Precio', ascending=True)

plt.figure(figsize=(10,6))
plt.barh(cmp_tienda.index, cmp_tienda['Precio'])
plt.title('Categorias mas populares de la Tienda 1')
plt.xlabel('Total por Ventas')
plt.ylabel('Categori')
plt.show()
```

```python
tienda['Fecha de Compra'] = pd.to_datetime(tienda['Fecha de Compra'], dayfirst=True)
tienda['Mes'] = tienda['Fecha de Compra'].dt.month

ventas_mes = tienda.groupby('Mes')['Precio'].sum()

plt.plot(ventas_mes)
plt.xlabel('Mes')
plt.ylabel('Ventas Totales')
plt.title('Ventas por Mes de la tienda 1')
plt.show()
```
